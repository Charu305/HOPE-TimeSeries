```python
performance={"Model":[],"RMSE":[],"MaPe":[],"Lag":[],"Test":[]}
```

```python
column="Close"
```

```python
def combination_ses(dataset, column):
    print("Running SES for:", column)
    series = dataset[column]
    test_obs = 28
    train = series[:-test_obs]
    test = series[-test_obs:]

    from statsmodels.tsa.holtwinters import SimpleExpSmoothing
    import pandas as pd
    import numpy as np
    from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

    # Fit SES model
    model = SimpleExpSmoothing(train)
    result = model.fit()

    # Forecast
    pred = result.forecast(steps=test_obs)
    preds = pd.DataFrame(pred, columns=[column])
    preds.to_csv("ses_forecasted_{}.csv".format(test_obs))

    # Evaluate
    rmse = np.sqrt(mean_squared_error(test, pred))
    mape = mean_absolute_percentage_error(test, pred)

    performance["Model"].append(column + "_SES")
    performance["RMSE"].append(rmse)
    performance["MaPe"].append(mape)
    performance["Lag"].append("N/A")
    performance["Test"].append(test_obs)

    perf = pd.DataFrame(performance)
    return perf, result, pred
```

```python
perf,result,pred=combination_ses(data1,column)
```

SES stands for Simple Exponential Smoothing.
It’s one of the simplest and most widely used time series forecasting methods — especially for data that:

✅ Has no clear trend,
✅ Has no seasonality,
✅ But may have random fluctuations.

⚙️ The basic idea

SES predicts the next value by taking a weighted average of all past observations, where recent values get more weight than older ones.

The weighting “fades exponentially” — meaning the influence of older observations decreases rapidly as you go further back in time.
